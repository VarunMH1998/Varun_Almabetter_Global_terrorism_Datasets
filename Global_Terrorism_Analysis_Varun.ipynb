# Global Terrorism Dataset Analysis

* Terrorism is a pressing global issue, and understanding its patterns and trends is crucial for effective policymaking, research, and public awareness. The "Global Terrorism Dataset Analysis" project is a comprehensive exploration of the Global Terrorism Database (GTD), an open-source repository containing information on terrorist events worldwide from 1970 to 2020. In this project, we employ data science techniques to uncover valuable insights, trends, and visualizations related to global terrorism.

# Introduction

* Terrorism remains a persistent and complex threat to global security. This project seeks to address the need for a deep understanding of this phenomenon by leveraging the Global Terrorism Database, one of the most comprehensive and reliable sources of information on terrorist activities worldwide. By analyzing this data, we aim to provide policymakers, researchers, and the public with a clearer picture of global terrorism trends and patterns.

# The Dataset

* The Global Terrorism Database, maintained by the National Consortium for the Study of Terrorism and Responses to Terrorism (START), is an invaluable resource for studying terrorism. It includes a wide range of information on terrorist incidents, such as the location, date, and type of attack, as well as details about the perpetrators and victims. To replicate our analysis, one can obtain the GTD from the START website and place it in the 'data' directory.

# Analysis Approach
Our analysis employs Python and various data science libraries to explore the GTD comprehensively. The project consists of a series of Jupyter notebooks in the 'notebooks' directory, each focusing on different aspects of terrorism analysis. Users can run these notebooks to generate insights and visualizations, but let's highlight a few key areas of exploration:

**Attack Types and Targets**: We delve into the various types of attacks (e.g., bombings, armed assaults) and the types of targets (e.g., civilians, military) to understand the tactics and motivations behind different terrorist groups.

**Group Analysis**: We analyze the involvement of various terrorist groups, their ideologies, and their activities over time. This sheds light on the ever-changing landscape of terrorism.

**Casualty Analysis**: Understanding the human cost of terrorism is essential. We investigate the number of casualties, including deaths and injuries, to gauge the severity of attacks.

# **Data cleaning**

* Filtering out required columns

```python
df = df[["iyear", "imonth", "iday", "country", "country_txt", "region", "region_txt", "provstate", "city", "latitude",
         "longitude", "location", "summary", "success", "attacktype1", "attacktype1_txt", "targtype1_txt",
         "target1", "natlty1", "natlty1_txt", "gname", "motive", "weaptype1", "weaptype1_txt", "nkill", "nwound", "addnotes"]]
```

```python
df.columns
```

* Renaming the columns to the meaningfull names

```python
df.rename(columns={"iyear":"year","imonth":"month", "iday":"day", "country":"country_id", "country_txt":"country", "region":"region_id",
                       "region_txt":"region", "provstate":"state", "attacktype1_txt":"attacktype", "targtype1_txt":"target",
                       "target1":"exact_target", "natlty1_txt":"nationality", "weaptype1_txt":"weapon_used", "nkill":"killed",
                       "nwound":"wounded"}, inplace=True)
```

```python
df.columns
```

```python
df.describe()
```

```python
df.reset_index()
```

* Finding the null values column wise

```python
df.isna().sum()
```

* Droping the duplicates

```python
df.drop_duplicates(inplace=True)
```

* After droping all the duplicate values, these are the total null values presentin each column

```python
df.isna().sum()
```

* Since there are no null values in killed and wounded columns, we can add these columns and make a new columns that is 'casuality', so that it will give sum of wounded and killed.

```python
df['killed'] = df['killed'].fillna(0)
df['wounded'] = df['wounded'].fillna(0)
df['casualty'] = df['killed']+df['wounded']
```

* We can find that new column casualty is created using wounded and killed columns

```python
df_casuality = df['casualty']
df_casuality
```

* we can find the sum of killed, wounded and casuality

```python
killed_sum = df['killed'].sum()
killed_sum
```

```python
wounded_sum = df['wounded'].sum()
wounded_sum
```

```python
casulity_sum = df['casualty'].sum()
casulity_sum
```

#Data Visualization

Yearwise attacks

```python
Top_attacks = attacks.sort_values(by='Attacks', ascending = False)
Top_attacks
```

```python
Top_attacks.head()
```

* We can conclude from the above graph that most and minimum attack between 1970 to 2017

-> **Which terrorists group has used most number of attacks in what type of attack**

```python
filter_data = df[df['gname'] != 'Unknown']
```

```python
grouped_data = filter_data.groupby(['gname', 'attacktype']).size().reset_index(name='Count')
grouped_data
```

```python
result = grouped_data.sort_values('Count', ascending=False).drop_duplicates(subset='gname')
result = result.head(10)
result
```

```python
plt.figure(figsize=(12, 6))
sns.set(style="whitegrid")
sns.barplot(x="attacktype", y="Count", hue="gname", data=result, ci=None)
plt.xticks(rotation=90)
plt.xlabel("attacktype")
plt.ylabel("Count")
plt.title("Terrorist Groups and Most Frequent Attack Types")
plt.legend(title="Terrorist Group", title_fontsize=15)
plt.tight_layout()
plt.show()
```

* Out of top 10 terrorists group 7 chose bombing/Explosions as the most favorable type of attacks
